# Shows a selected Number of Images

```python
import numpy as np
from matplotlib import pyplot as plt
import matplotlib
import glob
import random
%matplotlib inline
```

```python
# how many images to show
numberOfImages = 5
imagePath = "pokemon_image_data.npy"
```

```python
if imagePath == "":
    for filename in glob.glob("*.*"):
        if '.npy' in filename:
            img_array = np.load(filename)
            for x in range(numberOfImages):
                imageNr = random.randint(0,len(img_array))
                plt.figure()
                plt.imshow(img_array[imageNr], cmap="gray")
else:
    if '.npy' in imagePath:
        img_array = np.load(imagePath)
        for x in range(numberOfImages):
            imageNr = random.randint(0,len(img_array))
            plt.figure()
            plt.imshow(img_array[imageNr], cmap="gray")
    
```
